### **Lets look at some useful metrics**
- **Visit/Customer** is the total number of visits from our site over total number of customers in each method.
- **Sold/Visit** is the total number of sold items over the total number of visits from website we have gained.
- **Revenue/Sold** is the average revenue per the number of items we have sold.
- **Revenue/Customer** is the average revenue we have gained from each customer.

These metrics are calculated for the first 6 weeks in the table below. It is observable that "Email + Call" has the best gain considering these metrics. the problem with it is that the method is time consuming (not as time consuming as "Call" method). Considering the fact that call center agent is paid 18.5 $/Hr; in canada for example, [Check Indeed Please](https://ca.indeed.com/career/call-center-representative), and regarding the fact that we have spent 10 minutes in each consumer we can certainly conclude that the revenue gained per customer, 183.6, outweighs the cost we have spent. So, this method is certainly worth investing more on.
